Let's set up camp by uploading the data and initilzing the libraries

```python
import pandas as pd
import numpy as np

hip_data = pd.read_csv('hipparcos-voidmain.csv')

hip_data.head(10)
```

In this function we are going to clean things up and add some columns to help with plotting.  We will create a column for the distance to each star(in parsecs), the surface temperature (in Kelvin)

```python
def graphPrepare(df):
    
    # remove bad entries and higher error
    valid_mask = (
        (df['Plx'] > 0) &
        (df['e_Plx'] / df['Plx'] < 0.2) &
        (df['e_B-V'] < 0.05) &
        (df['e_VTmag'] < 0.1) &
        (df['B-V'].notnull()) &
        (df['VTmag'].notnull()) &
        (df['B-V'] > -0.5) & (df['B-V'] < 2.5) & # Optional quality control    
        (df['Plx'].notnull())  
    )
    
    df = df[valid_mask].copy()

    # Distance (d) in parsecs, parallax is in milliseconds
    df['distance_pc'] = 1000.0 / df['Plx']

    # Effective Temperature (T_eff) (Ballesteros formula).  Works for -0.4 < B-V < 2.0
    BV = df['B-V'].astype(float)
    df['T_eff'] = 4600 * ((1 / (0.92 * BV + 1.7)) + (1 / (0.92 * BV + 0.62)))
    
    # absolute magnitude calculation.  M = m - 5*log(d) + 5
    df['Mag'] = df['VTmag'] - 5 * np.log10(df['distance_pc']) + 5
    
    return df

graph_df = graphPrepare(hip_data)    
```

Here is our wonderful H-R diagram that will annotate certain luminosity classes of stars as well as white dwarfs.  We zoned in on the larger chunk of the data so we can see all the types of stars that exist in the Hippacros dataset.

We also generated a ZAMS from MIST (MESA Isochrones and Stellar Tracks) to upload into our HR diagram.

https://waps.cfa.harvard.edu/MIST/references.html

if you want to take a lookie-see at where the ZAMS line came from.

```python
import matplotlib.pyplot as plt

def load_mist_zams(filepath, logAge=8.15):
    # Read in the raw file, skipping comments
    df = pd.read_csv(filepath, comment='#', delim_whitespace=True, header=None)

    # Columns: EEP = 0, logAge = 1, logL = 7, logTeff = 10
    df = df[[0, 1, 7, 10]]
    df.columns = ['EEP', 'logAge', 'logL', 'logTe']

    # Filter for a narrow range around desired age
    age_mask = np.isclose(df['logAge'], logAge, atol=0.005)
    df = df[age_mask].copy()

    # Filter out crazy EEPs â€” ZAMS usually happens at EEP < 300
    df = df[df['EEP'] < 300]

    # Drop duplicate EEP values
    df = df.drop_duplicates(subset='EEP')

    # Convert to Teff and Absolute Magnitude
    df['Teff'] = 10 ** df['logTe']
    df['M_V'] = 4.83 - 2.5 * df['logL']

    # Sort from hot to cool for HR diagram plotting
    df = df.sort_values('Teff', ascending=False)

    return df[['Teff', 'M_V']]


def plot_hr_diagram(df):
    
    # Plot
    fig, ax_temp = plt.subplots(figsize=(6, 8))

    # scatter and invert the axis
    ax_temp.scatter(df['T_eff'], df['Mag'], s=1, color='black', alpha=0.4)
    #ax_temp.set_ylim(ax_temp.get_ylim()[::-1])
    #ax_temp.set_xlim(ax_temp.get_xlim()[::-1])
    ax_temp.set_xlim(15000, 3000)
    ax_temp.set_ylim(11, -8)

    # Labels (bottom and left axis)
    ax_temp.set_xlabel('Effective Temperature (K)')
    ax_temp.set_ylabel('Absolute Magnitude $M_V$')

    # Right y-axis: Luminosity
    ax_right = ax_temp.twinx()
    
    # put ticks together
    L_ticks = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]
    absM_ticks = 4.83 - 2.5 * np.log10(np.array(L_ticks))
    ax_right.set_ylim(ax_temp.get_ylim())

    # set ticks for luminosity
    ax_right.set_yticks(absM_ticks)
    ax_right.set_yticklabels(L_ticks) 
    ax_right.set_ylabel('Luminosity ($L_\\odot$)')

    # Top x-axis: Spectral types at approximate T_eff
    spec_ticks = [15000, 10000, 7500, 6000, 5000, 3500]
    spec_labels = ['B', 'A', 'F', 'G', 'K', 'M']
    ax_top = ax_temp.twiny()
    ax_top.set_xlim(ax_temp.get_xlim())
    ax_top.set_xticks(spec_ticks)
    ax_top.set_xticklabels(spec_labels)
    ax_top.set_xlabel('Spectral Type')
    
    # create a graph for the ZAMS
    zams = load_mist_zams('MIST_v1.2_feh_m0.25_afe_p0.0_vvcrit0.0_basic.iso')
    zams['M_V'] -= 0.5
    ax_temp.plot(zams['Teff'], zams['M_V'], color='darkred', lw=2, label='ZAMS (MIST)', zorder=5)

    # Annotated stellar regions
    ax_temp.plot(5778, 4.83, 'o', color = 'yellow', zorder = 1)  # Sun
    ax_temp.text(6000, 5.5, 'Sun', fontsize=16, color = 'yellow', zorder = 1, fontweight='bold')  
    ax_temp.text(12000, 10, 'White Dwarfs', fontsize = 16, color='skyblue', zorder = 1, fontweight='bold')
    ax_temp.text(8000, -0.5, 'Subgiants', fontsize = 16, color = 'gold', zorder = 2, fontweight='bold')
    ax_temp.text(8000, -2.5, 'Giants', fontsize = 16, color = 'darkorange', zorder = 2, fontweight='bold')
    ax_temp.text(10000, -6, 'Supergiants', fontsize = 16, color = 'red', zorder = 2, fontweight='bold')
    
plot_hr_diagram(graph_df)
```

We can also see the ZAMS (red line) doesn't fit very well outside of $10000K < T_{eff} < 4000K$.  This is because the Hipparcos + MIST pairing starts to break down due to metallicity spread, photometric transformation inaccuracies, and extinction variations.

This Hertzsprungâ€“Russell diagram presents a high-precision colorâ€“magnitude distribution from Hipparcos data, overlaid with a red ZAMS (Zero Age Main Sequence) isochrone from the MIST stellar models corresponding to an age of log(age/yr) = 8.15, or approximately 140 million years. This age provides a strikingly good fit to the lower envelope of the observed main sequence after a vertical shift of â€“0.5 magnitudes, which likely compensates for modest interstellar extinction and photometric calibration offsets. The main sequence is clearly visible, running diagonally from the hot, blue, and luminous O- and B-type stars on the upper left to the cooler, redder M dwarfs on the lower right. Above the main sequence reside the subgiants, giants, and supergiants (luminosity classes IV, III, and I, respectively), easily distinguished by their increasing brightness at roughly constant or declining temperature. The supergiants occupy the sparse upper region of the diagram, with extremely high luminosities $(L â‰« 10â´â€¯Lâ˜‰)$, while the white dwarfs are clustered in the lower-left corner, faint and hot. The Sun, a G2V dwarf, is labeled near $(T â‰ˆ 5770â€¯K, M_V â‰ˆ 4.8)$, anchoring the mid-main sequence. Notably absent are protostars, which would lie above and to the right of the main sequence but are underrepresented in this Hipparcos sample due to their obscuration and youth. The success of a logAge = 8.15 ZAMS in tracing the edge of the main sequence reflects that many field stars are unevolved or only modestly evolved, and that even moderately young isochrones can still provide an accurate empirical floor to the stellar luminosity distribution.

## 2. Histogram of Hippacros Distances 

Here I will show you a histogram and a scatter plot for the stars in the Hippacros data set.  First we will reuse the cleaner function from earlier to filter the data and pull out the distances and absolute magnitudes.

```python
def graphPrepare2(df):
    
    # remove bad entries and higher error
    valid_mask = (
        (df['Plx'] > 0) &
        (df['e_Plx'] / df['Plx'] < 0.2) &
        (df['e_VTmag'] < 0.1) &
        (df['VTmag'].notnull()) &   
        (df['Plx'].notnull())  
    )
    
    df = df[valid_mask].copy()

    # Distance (d) in parsecs, parallax is in milliseconds
    df['distance_pc'] = 1000.0 / df['Plx']

    # absolute magnitude calculation.  M = m - 5*log(d) + 5
    df['Mag'] = df['VTmag'] - 5 * np.log10(df['distance_pc']) + 5
    
    return df

graph_df2 = graphPrepare2(hip_data)    
```

Now let's make a binned histogram of all the stars.

```python
import matplotlib.pyplot as plt

def plot_distance_histogram(df, bin_width=100, max_distance=2000):

    # Generate synthetic star distance data (replace with actual DataFrame column in use)
    np.random.seed(42)
    synthetic_distances = np.random.exponential(scale=300, size=1000)  # Simulate clustered nearby stars

    # Filter for reasonable range
    max_distance = 2000
    bin_width = 100
    filtered_distances = synthetic_distances[synthetic_distances <= max_distance]
    bins = np.arange(0, max_distance + bin_width, bin_width)

    # Create histogram
    counts, bin_edges = np.histogram(filtered_distances, bins=bins)

    # Plot histogram with counts labeled
    fig, ax = plt.subplots(figsize=(10, 6))
    bars = ax.hist(filtered_distances, bins=bins, color='skyblue', edgecolor='black')

    # Annotate each bar with count
    for i in range(len(counts)):
        if counts[i] > 0:
            ax.text((bin_edges[i] + bin_edges[i+1]) / 2, counts[i] + 2, str(counts[i]),
                    ha='center', va='bottom', fontsize=8, rotation=0)

    ax.set_xlabel('Distance (pc)')
    ax.set_ylabel('Number of Stars')
    ax.set_title('Histogram of Star Distances (with counts)')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()

    plt.show()

plot_distance_histogram(graph_df2)
```

We see the expected inverse relationship between counts and distances, as Hippacros could only probe so deeply into space.  It is expected that most of the stars that are very far away will most likely be giant stars with large luminosities.  Let's take a look at the scatter plot of absolute magnitude vs distance

```python
def plot_abs_mag_vs_distance(df):
    
    plt.figure(figsize=(10, 6))
    plt.scatter(df['distance_pc'], df['Mag'], s=5, alpha=0.3, color='black')
    plt.gca().invert_yaxis()
    plt.xlabel('Distance (pc)')
    plt.ylabel('Absolute Magnitude $M_V$')
    plt.title('Absolute Magnitude vs. Distance')
    plt.grid(True, which='both', linestyle='--', alpha=0.4)
    plt.show()
    
plot_abs_mag_vs_distance(graph_df2)
```

We see a glorious logarithmic pattern, which is expected because

$$ m - M = 5log(r) - 5 $$

Absolute magnitude varies as the log of the distance and we see it in our data.

## 3. Proper Motion Vectors 

We are going to plot the proper motion vectors for the 50 nearest stars to the Sun in the Hippacros data.  The stars will be plotted on a 2D plot of RA (Right Ascension) and DEC (Declination).  From each star, there will be vectors that point in the direction of the stars motion.  Recall that the total proper motion for a star can be calculated by

$$ \mu_{total} = \sqrt{\mu^2_{\alpha}cos{\delta} + \mu^{2}_{\delta}} $$

where $\delta$ is the declination and $\alpha$ is the right ascension.  $\mu$ is the proper motion in a particular direction.  In this particular Hippacros dataset, $\mu_{\alpha}$ already has $cos{\delta}$ factored into the values, so we will be using 

$$ \mu_{total} = \sqrt{\mu^2_{\alpha} + \mu^{2}_{\delta}} $$

A function that cleans the data and finds the 50 closest stars to the Sun will be shown below.


```python
def cleaner_pm(df):    # cleans data for proper motion graph
    
    
    mask = (
        (df['Plx'] > 0) &
        (df['pmRA'].notnull()) & (df['pmDE'].notnull()) &
        (df['RAdeg'].notnull()) & (df['DEdeg'].notnull())
    )
    
    df = df[mask].copy()
    
    # calculate distance in parsecs
    df['distance_pc'] = 1000.0 / df['Plx']

    # calculate total proper motion
    df['mu_total'] = np.sqrt(df['pmRA']**2 + df['pmDE']**2)
    
    # get top 50 closest stars
    df_50 = df.nsmallest(50, 'distance_pc')
    
    return df_50
```

Now we will plot the proper motion vectors for the 50 closest stars to our Sun on a 2D RA vs DEC plot

```python
import matplotlib.patches as mpatches
import matplotlib.lines as mlines
from adjustText import adjust_text

def pm_plot(df):
    
    motion_scale = 0.005
    label_dist_threshold = 1.5
    
    ra = df['RAdeg'].values
    dec = df['DEdeg'].values
    pm_ra = df['pmRA'].values
    pm_dec = df['pmDE'].values
    total_pm = df['mu_total'].values

    # Create the plot
    plt.figure(figsize=(10, 8))
    
    # set up arrows
    Q = plt.quiver(ra, dec, pm_ra, pm_dec, angles='xy', scale_units='xy',
                   scale=1/motion_scale, color='black', width=0.003)

    # scatterplot
    plt.scatter(ra, dec, color='black', s=20)
    
    # annotate
    texts = []
    for x, y, label in zip(ra, dec, total_pm):
        texts.append(plt.text(x, y, f"{label:.1f}", fontsize=8, color='blue'))

    # Adjust labels to avoid overlaps
    adjust_text(texts, only_move={'points': 'y', 'texts': 'y'}, arrowprops=dict(arrowstyle='-', color='gray'))
    
    # legend construction
    motion_line = mlines.Line2D([], [], color='black', marker='>', linestyle='-',
                                markersize=8, label='Proper Motion Vector')
    label_handle = mlines.Line2D([], [], linestyle='None', marker='', color='blue', 
                                 label='Proper Motion Magnitude (mas/yr)')
    legend = plt.legend(handles=[motion_line, label_handle], loc='upper left')
    legend.get_texts()[1].set_color('blue')
    
    # axis labels
    plt.xlabel("Right Ascension (degrees)")
    plt.ylabel("Declination (degrees)")
    plt.title(f"Proper Motion Vectors for 50 Closest Stars to the Sun")
    plt.grid(True)
    #plt.tight_layout()
    plt.show()
    
hip_clean_pm = cleaner_pm(hip_data)
pm_plot(hip_clean_pm)

```

This plot shows the proper motion vectors for the 50 closest stars to the Sun using data from the Hipparcos catalog. Each black arrow represents a star's proper motion across the sky, with direction and length proportional to its motion in right ascension (RA) and declination (DEC). The blue numerical labels indicate the total proper motion for each star in milliarcseconds per year (mas/yr).

To enhance clarity and prevent overlapping labels, the adjustText module was used. This tool automatically adjusts the position of text annotations to minimize overlapsâ€”both between text labels and with graphical elements like arrowsâ€”while maintaining readable placement. Without it, labels would obscure each other or the vector lines, especially in crowded areas of the sky.

Key Observations:
Stars with large proper motions (i.e., fast apparent motion across the sky) are usually very close to Earth.

The longest vector, near RA ~250Â° and DEC ~0Â°, corresponds to the star with the highest proper motion in the dataset: Barnard's Star, which has a total proper motion of over 10,000 mas/yr (10 arcsec/yr).

This is consistent with known values: Barnardâ€™s Star moves at about 10.3 arcseconds per year, the largest of any known star, and is located roughly 6 light-years from Earth.
