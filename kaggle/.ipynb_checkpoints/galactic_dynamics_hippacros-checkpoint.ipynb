# Using Galactic Dynamics to Map Our Milky Way w/Hippacross

This notebook explores the 3D structure of the Milky Way using Hipparcos data, focusing on stellar motions and density across the Galactic disk.

We are going to select stars with errors that are less than 20 om "parallax" and the proper motion values 

```python
import pandas as pd
import numpy as np

def cleaner(df):

    # Filtering based on quality of measurements
    valid_mask = (
        (df['Plx'] > 0) &
        (df['e_Plx'] / df['Plx'] < 0.2) &
        (df['RAdeg'].notnull()) &
        (df['DEdeg'].notnull()) &
        (df['pmRA'].notnull()) &
        (df['pmDE'].notnull()) &
        (df['e_pmRA'] / df['pmRA'] < 0.2) &
        (df['e_pmDE'] / df['pmDE'] < 0.2)
    )
    
    cleaned_df = df[valid_mask].copy()

    # Calculate distance in parsecs from parallax in milliarcseconds
    cleaned_df['distance_pc'] = 1000.0 / cleaned_df['Plx']

    # Select relevant columns
    columns_needed = ['RAdeg', 'DEdeg', 'Plx', 'distance_pc', 'pmRA', 'pmDE']
    cleaned_df = cleaned_df[columns_needed]

    return cleaned_df

hip_data = pd.read_csv('hipparcos-voidmain.csv')

hip_data.head(10)

cleaned_df = cleaner(hip_data)
```

We are going to work on two plots: We will plot $\mu_{\delta}$ and $\mu_{\alpha}$

## 1. Proper Motion Vectors Over (b, l) 

This plot visualizes the proper motion vectors of stars in Galactic coordinates (longitude l and latitude b), projected onto the plane of the sky. Each arrow represents the direction and relative magnitude of a starâ€™s motion across the sky, based on its measured proper motion from the Hipparcos catalog. By examining these vectors, we can trace large-scale stellar motion patterns and identify the organized rotation of the Galactic disk or the random motion of halo stars.

```python
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import SkyCoord

def plot_proper_motion_vectors(df, max_arrows=1000):
   
    # Subsample for clarity
    sample_df = df.sample(min(len(df), max_arrows), random_state=15)
    
    # Get positions and proper motions
    coords = SkyCoord(ra=sample_df['RAdeg'].values * u.deg,
                      dec=sample_df['DEdeg'].values * u.deg,
                      distance=(1000 / sample_df['Plx'].values) * u.pc,
                      pm_ra_cosdec=sample_df['pmRA'].values * u.mas/u.yr,
                      pm_dec=sample_df['pmDE'].values * u.mas/u.yr)
    
    gal = coords.transform_to('galactic')
    l = gal.l.wrap_at(180 * u.deg).degree
    b = gal.b.degree
    pm_l = gal.pm_l_cosb.value
    pm_b = gal.pm_b.value

    plt.figure(figsize=(10, 10))
    plt.quiver(l, b, pm_l, pm_b, scale=2500, width=0.002, color='blue', alpha=0.6)
    plt.xlabel('Galactic Longitude (l)')
    plt.ylabel('Galactic Latitude (b)')
    plt.title('Proper Motion Vectors in Galactic Coordinates')
    plt.grid(True)
    plt.show()
    
plot_proper_motion_vectors(cleaned_df)
```

This plot shows the proper motion vectors of stars from the Hipparcos catalog in Galactic coordinates. The dense clustering around the Galactic equator (b = 0) illustrates the structure of the Milky Wayâ€™s stellar disk. The systematic proper motion patterns reflect the Sunâ€™s motion relative to local stars and the intrinsic velocity dispersion of stars within the disk. This analysis provides a two-dimensional kinematic map that hints at the Galactic rotation and local streaming motions, even without radial velocities.

## 2.  3D Position Plot (X, Y, Z)

In the absence of radial velocities, we can still explore the spatial structure of the Milky Way using Hipparcos data by transforming star positions from equatorial coordinates (RA, Dec, Parallax) into Galactic Cartesian coordinates (X, Y, Z). This lets us map the three-dimensional distribution of stars in the local volume (typically within ~500 pc).

To distinguish Galactic components:

We color-code stars based on height above or below the Galactic plane (Z) and Galactic latitude (b):

Thin Disk: |Z| < 300 pc, low |b|

Thick Disk: 300 < |Z| < 1000 pc

Halo: |Z| > 1000 pc or high proper motion at high |b|

Bulge: Hard to detect with Hipparcos due to distance limits

We cannot resolve full 3D velocities (no radial velocity), but we can still use proper motion direction and position to infer general structural patterns and motion distributions.

First we define a function that adds [X,Y,Z] for each star given its l, b, and distance.

```python

def add_galactic_cartesian(df):

    # Create SkyCoord
    coords = SkyCoord(ra=df['RAdeg'].values * u.deg,
                      dec=df['DEdeg'].values * u.deg,
                      distance=(1000.0 / df['Plx'].values) * u.pc, frame='icrs')

    # Convert to Galactic
    gal = coords.galactic

    # Convert to Cartesian in Galactic frame
    cart = gal.cartesian

    # Store back into dataframe in parsecs
    df['X'] = cart.x.to(u.pc).value
    df['Y'] = cart.y.to(u.pc).value
    df['Z'] = cart.z.to(u.pc).value
    df['distance_pc'] = coords.distance.to(u.pc).value

    return df

df_coord = add_galactic_cartesian(cleaned_df)
```

Then we prepare our 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def plot_galactic_structure_3d(df, distance_limit_pc=200):

    # Compute distances in parsecs
    df = df[df['distance_pc'] < distance_limit_pc]

    # Classify stars
    def classify_z(z):
        if abs(z) < 300:
            return 'Thin Disk'
        elif abs(z) < 1000:
            return 'Thick Disk'
        else:
            return 'Halo'

    df['component'] = df['Z'].apply(classify_z)

    # Color map
    colors = {'Thin Disk': 'cornflowerblue', 'Thick Disk': 'orange', 'Halo': 'gray'}

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')

    for comp in df['component'].unique():
        subset = df[df['component'] == comp]
        ax.scatter(subset['X'], subset['Y'], subset['Z'],
                   c=colors[comp], label=comp, s=3, alpha=0.6)

    ax.set_xlabel('X (pc)')
    ax.set_ylabel('Y (pc)')
    ax.set_zlabel('Z (pc)')
    ax.set_title('3D Spatial Distribution of Stars (Hipparcos)')
    ax.legend()
    plt.tight_layout()
    plt.show()
    
plot_galactic_structure_3d(df_coord)
```

This 3D map reveals the strong planar concentration of stars in our Galaxyâ€™s thin disk within 500 pc, with very few stars far above or below the plane. It highlights the Sunâ€™s location in the disk and gives a clear visual impression of the Milky Wayâ€™s local structure.

## 3. XZ Plane Projection 

This plot shows a side-on (edge-on) slice of the Milky Way by plotting stars in the X-Z plane. It highlights how most stars are concentrated near the Galactic plane (Z=0), forming a thin disk, with fewer stars found at greater heights above or below the plane. This visualization provides an intuitive look at the vertical thickness of the local disk and reveals how it flares outward.

```python

def plot_xz_projection(df, distance_limit_pc=500):
    df_sub = df[df['distance_pc'] < distance_limit_pc]
    plt.figure(figsize=(8,6))
    plt.scatter(df_sub['X'], df_sub['Z'], s=1, alpha=0.5, color='navy')
    plt.xlabel('X (pc)')
    plt.ylabel('Z (pc)')
    plt.title('X vs Z Projection (Side View of Disk)')
    plt.grid(True)
    plt.show()
    
plot_xz_projection(df_coord) 
```

This edge-on projection of the Milky Wayâ€™s local stellar population clearly shows a flattened, disk-like distribution. Most stars cluster tightly around the Galactic midplane (Z â‰ˆ 0), with the density falling off smoothly above and below it. This is exactly what we expect from our position embedded in the Galactic thin disk â€” a structure dominated by stars orbiting in a relatively thin layer around the center of the Milky Way.

There is also a slight vertical spread, with stars reaching several hundred parsecs above and below the plane. This is consistent with the known presence of a thicker disk component and some local contamination from older stars in higher orbits (and perhaps halo interlopers).

Our next diagram is a histogram that illustrates the distribution of stars as a function of their vertical distance (Z) from the Galactic plane. It shows a strong central peak, indicating that most stars lie close to the mid-plane, with the number of stars falling off sharply at higher |Z|. This provides a clear statistical view of the diskâ€™s scale height and allows us to identify the transition toward thicker disk or halo populations.

```python

def plot_z_histogram(df, distance_limit_pc=500):
    df_sub = df[df['distance_pc'] < distance_limit_pc]
    plt.figure(figsize=(8,6))
    plt.hist(df_sub['Z'], bins=60, color='teal', alpha=0.7)
    plt.xlabel('Z (pc)')
    plt.ylabel('Number of Stars')
    plt.title('Distribution of Stars Above and Below Galactic Plane')
    plt.grid(True)
    plt.show()
    
plot_z_histogram(df_coord)
```

This histogram shows the number of stars as a function of their height (Z) above or below the Galactic plane, effectively slicing through the Galaxy vertically. The distribution is strongly peaked at Z â‰ˆ 0â€¯pc â€” right at the plane â€” and falls off symmetrically on both sides. This is exactly the signature of a thin stellar disk, with most stars concentrated close to the midplane.

The shape resembles a Gaussian or exponential falloff, reflecting how the stellar density decreases with distance from the Galactic midplane. Such profiles are well-studied and are used to define the diskâ€™s scale height (roughly how far you must move vertically before the density drops by a factor of e).
