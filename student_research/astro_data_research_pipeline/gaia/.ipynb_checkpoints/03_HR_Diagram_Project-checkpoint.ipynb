```python
## Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Choose a Target Region

We will query a sky region using a cone search.

You will set:
- RA (deg)
- Dec (deg)
- radius (deg)

Tip: Start with radius = 1–3 degrees for a cluster-sized region.

```python
# Replace these with your target region (degrees)

ra0 = 56.75    # example: near Pleiades
dec0 = 24.12
radius_deg = 2.0
```

## Load Data (Cached OR Home Query)

At school, internet may be blocked.
At home, you can run the query and save the CSV for later.

```python

# Reupload .csv if the data is already downloaded

#csv_path = "gaia_hr_region.csv"  # change if needed
#df = pd.read_csv(csv_path)
#df.head()

# HOME ONLY (requires internet)
# Uncomment to run at home

from astroquery.gaia import Gaia

query = f"""
SELECT
     source_id,
     ra, dec,
     parallax, parallax_error,
     pmra, pmra_error,
     pmdec, pmdec_error,
     phot_g_mean_mag,
     phot_bp_mean_mag,
     phot_rp_mean_mag,
     ruwe
 FROM gaiadr3.gaia_source
 WHERE
     CONTAINS(
         POINT('ICRS', ra, dec),
         CIRCLE('ICRS', {ra0}, {dec0}, {radius_deg})
     ) = 1
     AND phot_g_mean_mag IS NOT NULL
     AND phot_bp_mean_mag IS NOT NULL
     AND phot_rp_mean_mag IS NOT NULL
     AND parallax IS NOT NULL
"""

job = Gaia.launch_job_async(query)
results = job.get_results()
df = results.to_pandas()
df.to_csv("gaia_hr_region.csv", index=False)
df.head()
```

## Compute Color and Distance Proxy

We compute:
- Color index: BP - RP
- Parallax S/N: parallax_over_error
- Distance proxy: dist_pc_proxy = 1000/parallax (pc)

Distance proxy is only safe when parallax S/N is decent.

```python
df = df.copy()

df["bp_rp"] = df["phot_bp_mean_mag"] - df["phot_rp_mean_mag"]
df["parallax_over_error"] = df["parallax"] / df["parallax_error"]

# Distance proxy (pc); keep in mind: only meaningful for decent parallaxes
df["dist_pc_proxy"] = 1000.0 / df["parallax"]

df[["bp_rp", "phot_g_mean_mag", "parallax_over_error", "dist_pc_proxy"]].head()
```

## Raw Observed CMD

Observed CMD:
- x-axis: BP - RP
- y-axis: G magnitude

This includes foreground/background stars and measurement noise.

```python
plt.figure()
plt.scatter(df["bp_rp"], df["phot_g_mean_mag"], s=4)
plt.gca().invert_yaxis()
plt.xlabel("BP - RP (mag)")
plt.ylabel("G (mag)")
plt.title("Observed CMD (Raw)")
plt.show()
```

## Apply Basic Quality Cuts

We apply intro-level cuts to reduce noise:

- finite values
- parallax_over_error > 5 (or 10 for stricter sample)
- RUWE < 1.4 (astrometric sanity check)

These cuts will improve the clarity of the CMD/HR diagram.

```python
clean = df.copy()

needed = ["bp_rp", "phot_g_mean_mag", "parallax", "parallax_error", "ruwe"]
for col in needed:
    clean = clean[np.isfinite(clean[col])]

clean = clean.replace([np.inf, -np.inf], np.nan).dropna(subset=needed)

# Recompute S/N after dropping NaNs
clean["parallax_over_error"] = clean["parallax"] / clean["parallax_error"]

# Thresholds (adjustable)
sn_cut = 5
ruwe_cut = 1.4

clean = clean[(clean["parallax_over_error"] > sn_cut) & (clean["ruwe"] < ruwe_cut)]

clean.shape
```

## CMD After Cleaning

Now we replot the CMD.
A successful clean sample will show a tighter main sequence.

```python
plt.figure()
plt.scatter(clean["bp_rp"], clean["phot_g_mean_mag"], s=4)
plt.gca().invert_yaxis()
plt.xlabel("BP - RP (mag)")
plt.ylabel("G (mag)")
plt.title(f"Observed CMD (Cleaned: S/N>{sn_cut}, RUWE<{ruwe_cut})")
plt.show()
```

## From Apparent to Absolute Magnitude (Approx HR Diagram)

To approximate an HR diagram, we compute **absolute magnitude**:

M_G = G - 5 * log10(d/10 pc)

where d is distance in parsecs.

We will use the distance proxy (1000/parallax) for this educational project.
This is only reasonable for good parallax S/N.

```python
# Use clean sample only
clean = clean.copy()

# distance in pc
clean["dist_pc_proxy"] = 1000.0 / clean["parallax"]

# distance modulus
clean["dist_mod"] = 5 * np.log10(clean["dist_pc_proxy"] / 10.0)

# absolute magnitude in Gaia G band
clean["M_G"] = clean["phot_g_mean_mag"] - clean["dist_mod"]

clean[["phot_g_mean_mag", "dist_pc_proxy", "M_G"]].head()
```

## HR Diagram (Color vs Absolute Magnitude)

This plot is closer to a true HR diagram:
- x-axis: BP - RP (color proxy for temperature)
- y-axis: M_G (absolute magnitude proxy for luminosity)

Note: This is still not perfect (no extinction correction, proxy distance),
but it is a meaningful step toward astrophysical interpretation.

```python
plt.figure()
plt.scatter(clean["bp_rp"], clean["M_G"], s=4)
plt.gca().invert_yaxis()
plt.xlabel("BP - RP (mag)")
plt.ylabel("M_G (mag)")
plt.title("Approx HR Diagram (Gaia: BP-RP vs Absolute G)")
plt.show()
```

## Sensitivity Test: How Do Cuts Change the HR Diagram?

A real research skill is checking if your result depends on your choices.

You will compare:
- S/N > 5 vs S/N > 10

```python

def make_clean(df, sn_cut=5, ruwe_cut=1.4):

    c = df.copy()
    
    c["bp_rp"] = c["phot_bp_mean_mag"] - c["phot_rp_mean_mag"]
    c = c.replace([np.inf, -np.inf], np.nan)
    
    needed = ["bp_rp", "phot_g_mean_mag", "parallax", "parallax_error", "ruwe"]
    
    c = c.dropna(subset=needed)
    c["parallax_over_error"] = c["parallax"] / c["parallax_error"]
    c = c[(c["parallax_over_error"] > sn_cut) & (c["ruwe"] < ruwe_cut)]
    
    c["dist_pc_proxy"] = 1000.0 / c["parallax"]
    c["dist_mod"] = 5 * np.log10(c["dist_pc_proxy"] / 10.0)
    c["M_G"] = c["phot_g_mean_mag"] - c["dist_mod"]
    
    return c

c5 = make_clean(df, sn_cut=5, ruwe_cut=1.4)
c10 = make_clean(df, sn_cut=10, ruwe_cut=1.4)

plt.figure()
plt.scatter(c5["bp_rp"], c5["M_G"], s=4)
plt.gca().invert_yaxis()
plt.xlabel("BP - RP (mag)")
plt.ylabel("M_G (mag)")
plt.title("Approx HR Diagram (S/N > 5)")
plt.show()

plt.figure()
plt.scatter(c10["bp_rp"], c10["M_G"], s=4)
plt.gca().invert_yaxis()
plt.xlabel("BP - RP (mag)")
plt.ylabel("M_G (mag)")
plt.title("Approx HR Diagram (S/N > 10)")
plt.show()
```
